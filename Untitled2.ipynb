```python
import cv2
import numpy as np
from ultralytics import YOLO
import cvzone
from datetime import datetime

# Dictionary of prices for different items
pprices = {
    'card': 60,
    'Apples': 20,
    'Banana': 15,
    'Milk': 50,
    'Bun': 69,
    'Cabbage': 45,
    'Cold Drink': 92,
    'Dry Fruit': 17,
    'Eggs': 88,
    'Grapes': 23,
    'Green Seeds': 77,
    'Oil': 12,
    'Pineapple': 34,
    'Snacks': 81,
    'Vegetables': 56,
    'Water Bottle': 28,
    'bell pepper': 95,
    'bread': 10,
    'carrot': 41,
    'cucumber': 62,
    'fasol': 71,
    'garlic': 37,
    'hot pepper': 19,
    'onion': 53,
    'peas': 99,
    'salad': 25,
    'tin': 65,
    'tomato': 48
}

class SmartTrolley:
    def __init__(self):
        self.tracked_items = {}  # {track_id: {'item': name, 'price': price, 'last_seen': frame_count, 'position': (x, y)}}
        self.item_quantities = {}  # {item_name: quantity}
        self.total_bill = 0  # Initialize total_bill to 0
        self.frame_count = 0
        self.confidence_threshold = 0.4
        self.removal_threshold = 30  # Frames before considering item removed
        self.position_threshold = 50  # Pixel distance threshold for considering same item

    def calculate_distance(self, pos1, pos2):
        """Calculate distance between two positions"""
        return np.sqrt((pos1[0] - pos2[0])**2 + (pos1[1] - pos2[1])**2)

    def update_item_quantity(self, item_name, change):
        """Update quantity of an item"""
        if item_name not in self.item_quantities:
            self.item_quantities[item_name] = 0
        self.item_quantities[item_name] += change

    def remove_old_items(self):
        """Remove items that haven't been seen for a while"""
        items_to_remove = []
        for track_id, data in self.tracked_items.items():
            if self.frame_count - data['last_seen'] > self.removal_threshold:
                items_to_remove.append(track_id)
                self.total_bill -= data['price']
                self.update_item_quantity(data['item'], -1)
                print(f"Removed {data['item']} (ID: {track_id}) - Price: ${data['price']}")

        for track_id in items_to_remove:
            del self.tracked_items[track_id]

# Initialize the system
trolley = SmartTrolley()

# Load the YOLO model
model = YOLO("/home/alliance/Desktop/charan/runs/detect/train6/weights/best.pt")

# Initialize video capture
cap = cv2.VideoCapture("/home/alliance/Desktop/charan/Demo10_Retail1.mp4")  # Use 0 for webcam or provide video file path

while True:
    ret, frame = cap.read()
    if not ret:
        break

    trolley.frame_count += 1
    frame = cv2.resize(frame, (1020, 600))

    # Run YOLO tracking on the frame
    results = model.track(frame, persist=True,conf=0.8)

    # Process detection results
    if results[0].boxes is not None and results[0].boxes.id is not None:
        boxes = results[0].boxes.xyxy.int().cpu().tolist()
        class_ids = results[0].boxes.cls.int().cpu().tolist()
        track_ids = results[0].boxes.id.int().cpu().tolist()
        confidences = results[0].boxes.conf.cpu().tolist()

        current_frame_detections = set()  # Track detections in current frame

        for box, class_id, track_id, conf in zip(boxes, class_ids, track_ids, confidences):
             if conf < trolley.confidence_threshold:
                continue

             item_name = model.names[class_id]
             x1, y1, x2, y2 = box
             center_position = ((x1 + x2) // 2, (y1 + y2) // 2)

            # Avoid processing the same track_id multiple times
             if track_id in current_frame_detections:
                continue

             current_frame_detections.add(track_id)
             
              # Debugging: Print item name and check in pprices
             print(f"Detected item: {item_name}, Price from pprices: {pprices.get(item_name, 'Not found')}")

             # New item added to trolley
             if track_id not in trolley.tracked_items and item_name in pprices:
                trolley.tracked_items[track_id] = {
                    'item': item_name,
                    'price': pprices[item_name],
                    'last_seen': trolley.frame_count,
                    'position': center_position
                 }
                trolley.total_bill += pprices[item_name]
                trolley.update_item_quantity(item_name, 1)
                print(f"Added {item_name} (ID: {track_id}) - Price: ${pprices[item_name]}")
             elif track_id not in trolley.tracked_items: # Debugging for items not in pprices
                print(f"Warning: item {item_name} not found in price list")

             # Update tracking info for existing items
             if track_id in trolley.tracked_items:
                trolley.tracked_items[track_id]['last_seen'] = trolley.frame_count
                trolley.tracked_items[track_id]['position'] = center_position

             # Draw bounding box and label
             color = (0, 255, 0) if item_name == 'fivestar' else (0, 165, 255)
             cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

            # Add item label with tracking ID
             label = f"{item_name} (ID: {track_id})"
             cvzone.putTextRect(frame, label, (x1, y1-20), 1, 1)


    # Display total bill and quantities
    y_pos = 30
    cvzone.putTextRect(frame, f"Total Bill: ${trolley.total_bill:.2f}", (10, y_pos), 2, 2)

    y_pos += 40
    cvzone.putTextRect(frame, "Items in Trolley:", (10, y_pos), 2, 2)

    # Display quantities
    for item_name, quantity in trolley.item_quantities.items():
        if quantity > 0:
            y_pos += 30
            text_color = (0, 255, 0) if item_name == 'fivestar' else (255, 255, 255)
            cvzone.putTextRect(frame, f"{item_name}: x{quantity}", (10, y_pos), 1, 1)

    # Show the frame
    cv2.imshow("Smart Trolley System", frame)

    # Key controls
    key = cv2.waitKey(0) & 0xFF
    if key == ord('q'):
        break
    elif key == ord('r'):
        trolley = SmartTrolley()
        print("System reset!")
        trolley.total_bill = 0 
    

# Print final item summary at the end
print("Final Items in Trolley:")
for item_name, quantity in trolley.item_quantities.items():
    print(f"{item_name}: x{quantity}")

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
import cvzone
from datetime import datetime

# Dictionary of prices for different items
pprices = {
    'card': 60,
    'Apples': 20,
    'Banana': 15,
    'Milk': 50,
    'Bun': 69,
    'Cabbage': 45,
    'Cold Drink': 92,
    'Dry Fruit': 17,
    'Eggs': 88,
    'Grapes': 23,
    'Green Seeds': 77,
    'Oil': 12,
    'Pineapple': 34,
    'Snacks': 81,
    'Vegetables': 56,
    'Water Bottle': 28,
    'bell pepper': 95,
    'bread': 10,
    'carrot': 41,
    'cucumber': 62,
    'fasol': 71,
    'garlic': 37,
    'hot pepper': 19,
    'onion': 53,
    'peas': 99,
    'salad': 25,
    'tin': 65,
    'tomato': 48
}

class SmartTrolley:
    def __init__(self):
        self.tracked_items = {}  # {track_id: {'item': name, 'price': price, 'last_seen': frame_count, 'position': (x, y)}}
        self.item_quantities = {}  # {item_name: quantity}
        self.total_bill = 0  # Initialize total_bill to 0
        self.frame_count = 0
        self.confidence_threshold = 0.4
        self.removal_threshold = 30  # Frames before considering item removed
        self.position_threshold = 50  # Pixel distance threshold for considering same item

    def calculate_distance(self, pos1, pos2):
        """Calculate distance between two positions"""
        return np.sqrt((pos1[0] - pos2[0])**2 + (pos1[1] - pos2[1])**2)

    def update_item_quantity(self, item_name, change):
        """Update quantity of an item"""
        if item_name not in self.item_quantities:
            self.item_quantities[item_name] = 0
        self.item_quantities[item_name] += change

    def remove_old_items(self):
        """Remove items that haven't been seen for a while"""
        items_to_remove = []
        for track_id, data in self.tracked_items.items():
            if self.frame_count - data['last_seen'] > self.removal_threshold:
                items_to_remove.append(track_id)
                self.total_bill -= data['price']
                self.update_item_quantity(data['item'], -1)
                print(f"Removed {data['item']} (ID: {track_id}) - Price: ${data['price']}")

        for track_id in items_to_remove:
            del self.tracked_items[track_id]

# Initialize the system
trolley = SmartTrolley()

# Load the YOLO model
model = YOLO("/home/alliance/Desktop/charan/runs/detect/train10/weights/best.pt")

# Initialize video capture
cap = cv2.VideoCapture("/home/alliance/Desktop/charan/Demo10_Retail1.mp4")  # Use 0 for webcam or provide video file path

# Get video properties for saving
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'mp4v') # Use mp4v codec
output_path = "/home/alliance/Desktop/charan/Retail-Store-1/output5.mp4"
out = cv2.VideoWriter(output_path, fourcc, fps, (1020, 600))

while True:
    ret, frame = cap.read()
    if not ret:
        break

    trolley.frame_count += 1
    frame = cv2.resize(frame, (1020, 600))

    # Run YOLO tracking on the frame
    results = model.track(frame, persist=True)

    # Process detection results
    if results[0].boxes is not None and results[0].boxes.id is not None:
        boxes = results[0].boxes.xyxy.int().cpu().tolist()
        class_ids = results[0].boxes.cls.int().cpu().tolist()
        track_ids = results[0].boxes.id.int().cpu().tolist()
        confidences = results[0].boxes.conf.cpu().tolist()

        current_frame_detections = set()  # Track detections in current frame

        for box, class_id, track_id, conf in zip(boxes, class_ids, track_ids, confidences):
             if conf < trolley.confidence_threshold:
                continue

             item_name = model.names[class_id]
             x1, y1, x2, y2 = box
             center_position = ((x1 + x2) // 2, (y1 + y2) // 2)

            # Avoid processing the same track_id multiple times
             if track_id in current_frame_detections:
                continue

             current_frame_detections.add(track_id)
             
              # Debugging: Print item name and check in pprices
             print(f"Detected item: {item_name}, Price from pprices: {pprices.get(item_name, 'Not found')}")

             # New item added to trolley
             if track_id not in trolley.tracked_items and item_name in pprices:
                trolley.tracked_items[track_id] = {
                    'item': item_name,
                    'price': pprices[item_name],
                    'last_seen': trolley.frame_count,
                    'position': center_position
                 }
                trolley.total_bill += pprices[item_name]
                trolley.update_item_quantity(item_name, 1)
                print(f"Added {item_name} (ID: {track_id}) - Price: ${pprices[item_name]}")
             elif track_id not in trolley.tracked_items: # Debugging for items not in pprices
                print(f"Warning: item {item_name} not found in price list")

             # Update tracking info for existing items
             if track_id in trolley.tracked_items:
                trolley.tracked_items[track_id]['last_seen'] = trolley.frame_count
                trolley.tracked_items[track_id]['position'] = center_position

             # Draw bounding box and label
             color = (0, 255, 0) if item_name == 'fivestar' else (0, 165, 255)
             cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

            # Add item label with tracking ID
             label = f"{item_name} (ID: {track_id})"
             cvzone.putTextRect(frame, label, (x1, y1-20), 1, 1)


    # Display total bill and quantities
    y_pos = 30
    cvzone.putTextRect(frame, f"Total Bill: ${trolley.total_bill:.2f}", (10, y_pos), 2, 2)

    y_pos += 40
    cvzone.putTextRect(frame, "Items in Trolley:", (10, y_pos), 2, 2)

    # Display quantities
    for item_name, quantity in trolley.item_quantities.items():
        if quantity > 0:
            y_pos += 30
            text_color = (0, 255, 0) if item_name == 'fivestar' else (255, 255, 255)
            cvzone.putTextRect(frame, f"{item_name}: x{quantity}", (10, y_pos), 1, 1)

    # Show the frame
    cv2.imshow("Smart Trolley System", frame)

    # Write frame to video
    out.write(frame)
    
    # Key controls
    key = cv2.waitKey(0) & 0xFF
    if key == ord('q'):
        break
    elif key == ord('r'):
        trolley = SmartTrolley()
        print("System reset!")
        trolley.total_bill = 0  # Reset the total bill

# Print final item summary at the end
print("Final Items in Trolley:")
for item_name, quantity in trolley.item_quantities.items():
    print(f"{item_name}: x{quantity}")

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import os
import random
import shutil
import yaml
from pathlib import Path
import logging
from typing import List, Dict, Union

class DatasetSplitter:
    def __init__(self, config: Dict[str, Union[str, float]]):
        self.vcr_path = Path(config['source_path'])
        self.dataset_path = Path(config['dataset_path'])
        self.split_ratios = {
            'train': config['train_ratio'],
            'val': config['val_ratio'],
            'test': config['test_ratio']
        }
        self.image_extensions = config.get('image_extensions', ['.jpg', '.jpeg', '.png'])

        logging.basicConfig(
            level=logging.INFO,
            format='%(asctime)s - %(levelname)s - %(message)s'
        )
        self.logger = logging.getLogger(__name__)

    def setup_directories(self) -> None:
        """Create necessary directories for dataset splits."""
        for split in ['train', 'valid', 'test']:
            for subdir in ['images', 'labels']:
                path = self.dataset_path / split / subdir
                path.mkdir(parents=True, exist_ok=True)
                self.logger.info(f"Created directory: {path}")

    def collect_images(self) -> List[Path]:
        """Collect all valid image files from source directory."""
        images = []
        for ext in self.image_extensions:
            images.extend(list(self.vcr_path.glob(f'*{ext}')))

        if not images:
            raise FileNotFoundError(f"No images found in {self.vcr_path}")

        self.logger.info(f"Found {len(images)} images")
        return images

    def split_data(self, images: List[Path]) -> Dict[str, List[Path]]:
        """Shuffle and split data into train, validation, and test sets."""
        random.shuffle(images)  # Shuffle the dataset for randomness
        total = len(images)
        splits = {}

        start_idx = 0
        for split, ratio in self.split_ratios.items():
            count = int(total * ratio)
            splits[split] = images[start_idx:start_idx + count]
            start_idx += count
            self.logger.info(f"{split} set: {len(splits[split])} images")

        return splits

    def get_unique_classes(self) -> List[str]:
        """Extract unique class names from label files."""
        unique_classes = set()
        label_files = list(self.dataset_path.rglob('*.txt'))

        for label_file in label_files:
            try:
                with open(label_file, 'r') as f:
                    for line in f:
                        class_idx = line.strip().split()[0]
                        unique_classes.add(class_idx)
            except Exception as e:
                self.logger.warning(f"Error reading label file {label_file}: {str(e)}")
                continue

        return sorted(list(unique_classes))

    def create_yaml(self) -> None:
        """Create YAML configuration file for the dataset."""
        unique_classes = self.get_unique_classes()
        num_classes = len(unique_classes)

        yaml_content = {
            'path': str(self.dataset_path.absolute()),
            'train': str(self.dataset_path / 'train' / 'images'),
            'val': str(self.dataset_path / 'valid' / 'images'),
            'test': str(self.dataset_path / 'test' / 'images'),
            'nc': num_classes,
            'names': unique_classes
        }

        yaml_path = self.dataset_path / 'data.yaml'
        try:
            with open(yaml_path, 'w') as f:
                yaml.safe_dump(yaml_content, f, default_flow_style=False, sort_keys=False)
            self.logger.info(f"Created dataset configuration with {num_classes} classes: {yaml_path}")
        except Exception as e:
            self.logger.error(f"Error creating YAML file: {str(e)}")
            raise

    def move_files(self, split_name: str, files: List[Path]) -> None:
        """Move image and label files to their respective directories."""
        dest_split_name = 'valid' if split_name == 'val' else split_name
        dest_base = self.dataset_path / dest_split_name

        for img_path in files:
            # Copy image
            dest_img = dest_base / 'images' / img_path.name
            shutil.copy2(img_path, dest_img)

            # Copy label if exists
            label_path = img_path.with_suffix('.txt')
            if label_path.exists():
                dest_label = dest_base / 'labels' / label_path.name
                shutil.copy2(label_path, dest_label)
            else:
                self.logger.warning(f"No label file found for {img_path.name}")

    def process(self) -> None:
        """Execute the complete dataset splitting process."""
        try:
            self.setup_directories()
            images = self.collect_images()
            splits = self.split_data(images)

            for split_name, files in splits.items():
                self.move_files(split_name, files)

            self.create_yaml()
            self.logger.info("Dataset splitting completed successfully!")

        except Exception as e:
            self.logger.error(f"Error processing dataset: {str(e)}")
            raise

# Configuration
config = {
    'source_path': '/home/alliance/Desktop/charan/frames_1',  # Update this path
    'dataset_path': '/home/alliance/Desktop/charan/frames_2',  # Changed to match YOLO's expected path
    'train_ratio': 0.7,
    'val_ratio': 0.2,
    'test_ratio': 0.1,
    'image_extensions': ['.jpg', '.jpeg', '.png']
}

# Initialize and run DatasetSplitter
splitter = DatasetSplitter(config)
splitter.process()

# Verify the data.yaml contents
yaml_path = Path(config['dataset_path']) / 'data.yaml'
if yaml_path.exists():
    with open(yaml_path, 'r') as f:
        print("\nContents of data.yaml:")
        print(f.read())
else:
    print("\nError: data.yaml not found.")
 
```
